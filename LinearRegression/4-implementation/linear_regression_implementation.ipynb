```python
import numpy as np
import matplotlib.pyplot as plt
import linearregression as LRGD

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# generate random data-set
np.random.seed(0)
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)
m = x.shape[0]
```

```python
model = LRGD.LinearRegressionUsingGD()
x_train = np.c_[np.ones((m, 1)), x] #with 1 you find the intercept
fitting = model.fit(x_train, y)
prediction = model.predict(x_train)
cost_function = model.cost_
fig, ax = plt.subplots(1, 2, figsize=(12,4) )
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].scatter(x, y, s=5, color='tab:blue')
ax[0].plot(x, prediction, color='red', lw=0.05)
ax[1].set_title('Cost function')
ax[1].plot(cost_function)
ax[1].set_xlabel('steps')
ax[1].set_ylabel('Cost')
```

```python
##Evaluation of the Model
mse = np.sum((prediction - y)**2)
rmse = np.sqrt(mse/m)

ssr = np.sum((prediction - y)**2)
sst = np.sum((y - np.mean(y))**2)
r2_score = 1 - (ssr/sst)
print('The RMSE score is %0.10f\nSSr=%0.10f\nSSt=%0.10f\n' % (rmse, ssr, sst))
```

```python
#with sklearn
model = LinearRegression()
model.fit(x,y)
y_predicted = model.predict(x)
print('Intercept = %0.2f\nCoefficient = %0.2f' % (model.intercept_, model.coef_))

plt.plot(x, y_predicted, color='r', lw=0.5)
plt.scatter(x, y, s=5, color='tab:blue')
##Evaluation of the Model
rmse = np.sqrt(np.sum((y_predicted - y)**2)/m)
ssr = np.sum((y_predicted - y)**2)
sst = np.sum((y - np.mean(y))**2)
r2_score = 1 - (ssr/sst)
print('The RMSE score is %0.10f\nSSr=%0.10f\nSSt=%0.10f\n' % (rmse, ssr, sst))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
data = pd.read_csv('marks.txt', header = None)
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
admitted = data.loc[y == 1]
not_admitted = data.loc[y == 0]
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
theta = np.zeros((X.shape[1]))

model = LogisticRegression()
model.fit(X, y)
parameters = model.coef_
intercept = model.intercept_
print(intercept)
print(parameters)
parameters = np.squeeze(model.coef_, axis=0)

x_values = [np.min(X[0]-5), np.max(X[1]+5)]
y_values = -(intercept[0]+np.dot(parameters[-1], x_values))/parameters[-2]

plt.plot(x_values, y_values, label='Decision Boundary')
plt.xlim(10,110)
plt.ylim(10,110)
plt.scatter(admitted.iloc[:, 0], admitted.iloc[:, 1], s=10, label='Admitted')
plt.scatter(not_admitted.iloc[:, 0], not_admitted.iloc[:, 1], s=10, label='Not Admitted')
plt.legend()
```

```python
##evaluate predicted values with training data
predicted_classes = model.predict(X) 
accuracy = accuracy_score(y.to_numpy(), predicted_classes)
print(accuracy)
```
